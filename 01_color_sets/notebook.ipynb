```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors
```

```python
x = np.random.random(200)
y = np.random.random(200)
classes = np.random.randint(0, 3, 200)
```

```python
#Creating a custom DISCRETE colormap using matplotlib.colors.ListedColormap
custom_cmap_d = matplotlib.colors.ListedColormap(["red", "orange", "yellow"])

#Creating a custom SPECTRUM colormap using matplotlib.colors.ListedColormap
custom_cmap_s = matplotlib.colors.LinearSegmentedColormap.from_list("custom", ["red", "orange", "yellow"])
```

```python
plt.scatter(x, y, c=classes, cmap=custom_cmap_s)
plt.colorbar()
plt.show()
```

```python
import seaborn as sns
```

```python
data = sns.load_dataset("titanic")
```

```python
custom_cmap_s = matplotlib.colors.LinearSegmentedColormap.from_list("custom", ["#00aabb", "#ffffff", "#ff7777"])
```

```python
sns.heatmap(data.corr(), cmap=custom_cmap_s, annot=True, vmin=-1, vmax=1)
plt.show()
```
